```python
#Displays histogram of any numeric column selected
#plt.title adds a title to the histogram
#plt.xlabel and ylabel add labels to both x and y on histogram
df['math score'].hist()
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Number of Students")
```

```python
#Shows how to add a new column and to the right of the = shows what you want added in that column
#Two brackets are for more than one column
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
```

```python
#Making another new column 
df['passed'] = df['average score'] >= 60
df['passed'].value_counts()
```
